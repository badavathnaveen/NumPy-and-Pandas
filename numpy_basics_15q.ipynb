```python
'''
1. Create a numpy arrays of shape (3,) and concatenate across the column axis?
'''
x = np.array([1,2,3]).reshape((3,1))
y = np.array([4,5,6]).reshape((3,1))
x = np.concatenate((x,y),axis = 0)
x
```

```python
'''
2. Create a random numpy array of shape (50,) and filter out negative values?
'''
x = np.random.random(50)
filtered_x = x[x >= 0]
filtered_x
```

```python
'''
3. Create a random integer array of shape (100,) and normalize it?
'''
x = np.random.randint(1,100,size=100)
mean_x = np.mean(x)
centered_x = x-mean_x
std_x = np.std(x)
final_x = centered_x/std_x
x_final_mean = np.mean(final_x)
x_final_mean
```

```python
'''
13. Write a python code to create a list of numbers from 1 to 20 and convert them
    into two numpy arrays containing even and odd numbers?
'''
list1 = [i for i in range(1,20)]
b = np.array(list1)
x = b[b %2 == 0]
print("Even = ", x )
y = b[b % 2 != 0]
print("Odd = ",y)
```
